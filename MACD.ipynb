### Moving Average Convergence Divergence (MACD) 

In a bullish crossover, just like in Moving Averages, a buy signal occurs when MACD crosses above the signal line. 

A bearish signal occurs when MACD crosses below the signal line. 

If a crossover occurs with a high sloping MACD, this can be a sign of an overbought or oversold condition, depending on if the crossover is bullish or bearish respectively.

```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import pyEX as p

c = p.Client(api_token='pk_5f448ee7ba2f45cfb28974df0459c811')
ticker = 'AAPL'
timeframe = '6m'
df = c.chartDF(ticker, timeframe)
df = df[['close']]
df.reset_index(level=0, inplace=True)
df.columns=['ds','y']
plt.plot(df.ds, df.y, label='Apple')
plt.legend(loc='upper left')
plt.show()
```

A bullish crossover happens when the MACD crosses above the signal line and a bearish crossover happens when the MACD crosses below the signal line.

```python
exp1 = df.y.ewm(span=12, adjust=False).mean()
exp2 = df.y.ewm(span=26, adjust=False).mean()
macd = exp1-exp2
exp3 = macd.ewm(span=9, adjust=False).mean()
plt.plot(df.ds, macd, label='AAPL MACD', color = '#EBD2BE')
plt.plot(df.ds, exp3, label='Signal Line', color='magenta')
plt.legend(loc='upper left')
plt.show()
```

Exponential moving averages and MACD

```python
exp1 = df.y.ewm(span=12, adjust=False).mean()
exp2 = df.y.ewm(span=26, adjust=False).mean()
exp3 = df.y.ewm(span=9, adjust=False).mean()
macd = exp1-exp2
plt.plot(df.ds, df.y, label='AAPL')
plt.plot(df.ds, macd, label='AAPL MACD', color='orange')
plt.plot(df.ds, exp3, label='Signal Line', color='Magenta')
plt.legend(loc='upper left')
plt.show()
```

We can see the MACD stays pretty flat over time. But there are certain times where the MACD curve is steeper than others. 

These are instances of overbought or oversold conditions. 

You can see that soon after the MACD shows an overbought or oversold condition the momentum slowed and the stock price reacted accordingly.
